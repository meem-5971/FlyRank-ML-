## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the starter dataset directly from your repository's raw URL
raw_url = 'https://raw.githubusercontent.com/meem-5971/FlyRank-ML-/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(raw_url)

# Filter for baseline valid rows (matching the starter pipeline discipline)
valid_df = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)].drop_duplicates(subset=['content_id'])

# 1. Total volume of unique valid items to be clustered
total_items = len(valid_df)

# 2. Percentage of items sitting in a volatile position space (e.g., Page 1/2 fringe where archetypes diverge)
fringe_position_pct = (valid_df['avg_position'].between(5, 20).sum() / total_items) * 100

print(f"Number 1 (Total unique items to cluster): {total_items}")
print(f"Number 2 (Percentage of items in high-leverage position 5-20 fringe): {fringe_position_pct:.2f}%")
```
